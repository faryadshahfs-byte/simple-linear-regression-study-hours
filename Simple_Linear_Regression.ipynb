```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/drive/MyDrive/AI   ML 2026/student study hours - Sheet1 (2).csv")
df
```

```python
plt.scatter(df.Hours,df.Marks)
plt.xlabel("Hours", color="red")
plt.ylabel("Marks", color="green")
plt.title("Hours vs Marks")
plt.show()
```

```python
lr = linear_model.LinearRegression()
lr.fit(df[['Hours']],df.Marks)
```

```python
lr.predict([[9.25]])
```

```python
lr.score(df[['Hours']],df.Marks)
lr.coef_
```

```python
lr.intercept_
```
